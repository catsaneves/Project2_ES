```python
#drop columns in model of project 1
df_meteo2019=df_meteo2019.drop(columns=['HR', 'windSpeed_m/s', 'windGust_m/s', 'pres_mbar', 'rain_mm/h', 'rain_day'])
```

```python
df_meteo2019
```

```python
#putting 2019 data in the form of 2017 and 2018 data

df_power2019['Date']=pd.to_datetime(df_power2019['Date'],dayfirst=True)

df_power2019=df_power2019.set_index ('Date', drop = True)

df_power2019['Day']=df_power2019.index.date
```

```python
df_power2019
```

```python
df_power2019 = df_power2019.reset_index()

df_holiday2019=df_holiday2019.rename(columns={'Date':'Day'})

df_power_hol2019 = pd.merge(df_power2019, df_holiday2019, on='Day', how='left')
```

```python
df_power_hol2019
```

```python
df_power_hol2019=df_power_hol2019.set_index ('Date', drop = True)
df_power_hol2019['Weekday'] = pd.to_datetime(df_power_hol2019.index.values).weekday
df_power_hol2019['Hour'] = pd.to_datetime(df_power_hol2019.index.values).hour
df_power_hol2019 = df_power_hol2019.reset_index()
df_power_hol2019.head(30)
```

```python
#saturdays are a 5 and sundays are a 6
df_power_hol2019.loc[df_power_hol2019['Weekday']>=5, ['Holiday']]=1
df_power_hol2019.loc[df_power_hol2019['Weekday']<5, ['Holiday']]=0
#exam prep
df_power_hol2019.loc[(df_power_hol2019['Date'] > '2019-01-02') & (df_power_hol2019['Date'] < '2019-02-04') & (df_power_hol2019['Weekday'] < 5), ['Holiday']] = 2 
df_power_hol2019.loc[(df_power_hol2019['Date'] > '2019-06-02') & (df_power_hol2019['Date'] < '2019-06-07') & (df_power_hol2019['Weekday'] < 5), ['Holiday']] = 2  
#break
df_power_hol2019.loc[(df_power_hol2019['Date'] > '2019-02-05') & (df_power_hol2019['Date'] < '2019-02-18') & (df_power_hol2019['Weekday'] < 5), ['Holiday']] = 3
df_power_hol2019.loc[(df_power_hol2019['Date'] > '2019-04-14') & (df_power_hol2019['Date'] < '2019-04-20') & (df_power_hol2019['Weekday'] < 5), ['Holiday']] = 3
df_power_hol2019.loc[(df_power_hol2019['Date'] > '2019-07-10') & (df_power_hol2019['Date'] < '2019-09-09') & (df_power_hol2019['Weekday'] < 5), ['Holiday']] = 3
df_power_hol2019.loc[(df_power_hol2019['Date'] > '2019-12-20') & (df_power_hol2019['Date'] < '2020-01-01') & (df_power_hol2019['Weekday'] < 5), ['Holiday']] = 3

df_power_hol2019
```

```python
df_meteo2019['Date']=pd.to_datetime(df_meteo2019['Date'],dayfirst=True)
df_power_hol2019 = df_power_hol2019.reset_index()
df_2019 = pd.merge(df_power_hol2019, df_meteo2019, on='Date')
```

```python
df_2019
```

```python
df_2019['Power-1']=df_2019['Power_kW'].shift(1) # Previous hour consumption

df_2019=df_2019.drop(columns=['index'])
```

```python
df_2019.head()
```

```python
df_2019=df_2019.dropna()
df_2019
```

```python
df_2019=df_2019.drop(columns=['Day'])
df_2019=df_2019.set_index ('Date', drop = True)
#dataframe ready to be used for testing prediction methods
df_2019
```

```python
#will use data from 2017 and 2018 as train and 2019 as test
Z_1718=df_reg.values
#Identify output Y
Y_1718=Z_1718[:,0]
#Identify input Y
X_1718=Z_1718[:,[1,2,3,4,5,6]] 

Z_2019=df_2019.values
#Identify output Y
Y_2019=Z_2019[:,0]
#Identify input Y
X_2019=Z_2019[:,[1,2,3,4,5,6]] 
```

```python
from xgboost import XGBRegressor
from sklearn import  metrics

XGB_model19 = XGBRegressor()
XGB_model19.fit(X_1718, Y_1718)
y_pred_XGB =XGB_model19.predict(X_2019)
```

```python
plt.plot(Y_2019[1:200])
plt.plot(y_pred_XGB[1:200])
plt.show()
plt.scatter(Y_2019,y_pred_XGB)
```

```python
MAE_XGB=metrics.mean_absolute_error(Y_2019,y_pred_XGB) 
MSE_XGB=metrics.mean_squared_error(Y_2019,y_pred_XGB)  
RMSE_XGB= np.sqrt(metrics.mean_squared_error(Y_2019,y_pred_XGB))
cvRMSE_XGB=RMSE_XGB/np.mean(Y_2019)
comp_XGB=[MAE_XGB,MSE_XGB,RMSE_XGB,cvRMSE_XGB]
print(MAE_XGB,MSE_XGB,RMSE_XGB,cvRMSE_XGB)
```

```python
from sklearn.ensemble import BaggingRegressor

BT_model = BaggingRegressor()
BT_model.fit(X_1718, Y_1718)
y_pred_BT =BT_model.predict(X_2019)
```

```python
plt.plot(Y_2019[1:200])
plt.plot(y_pred_BT[1:200])
plt.show()
plt.scatter(Y_2019,y_pred_BT)
```

```python
MAE_BT=metrics.mean_absolute_error(Y_2019,y_pred_BT) 
MSE_BT=metrics.mean_squared_error(Y_2019,y_pred_BT)  
RMSE_BT= np.sqrt(metrics.mean_squared_error(Y_2019,y_pred_BT))
cvRMSE_BT=RMSE_BT/np.mean(Y_2019)
comp_BT=[MAE_BT,MSE_BT,RMSE_BT,cvRMSE_BT]
print(MAE_BT,MSE_BT,RMSE_BT,cvRMSE_BT)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

GB_model = GradientBoostingRegressor()
GB_model.fit(X_1718, Y_1718)
y_pred_GB =GB_model.predict(X_2019)
```

```python
plt.plot(Y_2019[1:200])
plt.plot(y_pred_GB[1:200])
plt.show()
plt.scatter(Y_2019,y_pred_GB)
```

```python
MAE_GB=metrics.mean_absolute_error(Y_2019,y_pred_GB) 
MSE_GB=metrics.mean_squared_error(Y_2019,y_pred_GB)  
RMSE_GB= np.sqrt(metrics.mean_squared_error(Y_2019,y_pred_GB))
cvRMSE_GB=RMSE_GB/np.mean(Y_2019)
comp_GB=[MAE_GB,MSE_GB,RMSE_GB,cvRMSE_GB]
print(MAE_GB,MSE_GB,RMSE_GB,cvRMSE_GB)
```
